```python
import pandas as pd
import numpy as np
import requests as rq
import time
from tqdm import tqdm_notebook as tn
from bs4 import BeautifulSoup as bs
from selenium import webdriver as wd
from selenium.webdriver.common.keys import Keys
```

```python
# 네이버 쇼핑 Selenium 호출
url = 'https://www.kurly.com/shop/goods/goods_search.php?searched=Y&sword='
driver = wd.Chrome()
driver.get(url)
```

```python
# 데이터프레임에서 검색할 상품명 가져왔던 코드

df = pd.read_csv("img_01.csv", encoding="utf-8")
for i in range(len(df)):
    # df['4'][i] = df['4'][i].replace(",","")
    df['PRODUCT_NAME'][i] = df['PRODUCT_NAME'][i].replace(" ","")
    df['PRODUCT_NAME'][i] = df['PRODUCT_NAME'][i].split(",")
a = set()
for i in df['PRODUCT_NAME']:
    a = a | set(i)
```

```python
## 데이터프레임에서 가져온 상품명들 list로 만들어주고 정렬한 코드

A = list(a)
A = sorted(A)
A
```

```python
img_list = []
for i in range(len(A)): # 한 페이지에 20개씩, 3개 페이지를 만들어야되니까 60번 돌도록
    search = driver.find_element_by_css_selector("#sword") #검색창 element 찾기
    time.sleep(1) # 찾았으면 쉬기
    search.send_keys(A[i]) # 검색창에 상품명 입력해주기
    search.send_keys(Keys.ENTER) # 검색
    time.sleep(3) # 이게 검색되고 시간이 좀 걸리더라고
    try :
        img = driver.find_element_by_css_selector("#goodsList > div.list_goods > div > ul > li:nth-child(1) > div > div > a > img")
        img_link = img.get_attribute("src")
        img_list.append(img_link) # 이미지 크롤링해서 img_list에 담아줬음
    except:
        img_link = "NONE" # 이미지가 없거나, 이미지 크롤링 되기 전 화면 넘어가면 걍 NONE
        img_list.append(img_link)

    time.sleep(1)
    search = driver.find_element_by_css_selector("#sword") # 검색창 다시 찾아줌
    search.send_keys(Keys.LEFT_SHIFT, Keys.HOME) # 검색했던 내용 다 지우는 코드 1
    search.send_keys(Keys.BACKSPACE) # 코드 2
```

```python
## 검색하고 item_list에다가 크롤링한거 담아준 코드

search = driver.find_element_by_css_selector('#__next > div > div.header_header__26kBA > div > div.gnb_header_shop__qMTrt > div > div.gnbSearch_search_area__1krSO > form > fieldset > div.gnbSearch_inner__1ptdl > input')

item_list = []
for i in range(len(A)):
    search.send_keys(Keys.LEFT_SHIFT, Keys.HOME)
    search.send_keys(Keys.BACKSPACE)
    time.sleep(1)
    search.send_keys(A[i])
    search.send_keys(Keys.ENTER)
    time.sleep(1.5)
    try:
        price = driver.find_element_by_css_selector('#__next > div > div.container > div > div.style_content_wrap__1PzEo > div > ul > div > div:nth-child(1)> li > div> div > div > strong > span')
        item_list.append(price.text)
    except:
        item_list.append('Fail')
    time.sleep(1.5)
```
